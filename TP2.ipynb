# Exercise 1: Explore Web Logs with Spark RDDs

```python
from pyspark.sql.types import *



#1
#schema = StructType([StructField('Project', StringType(), True),\
  #                   StructField('Page title', StringType(), True),\
  #                   StructField('Page hits', IntegerType(), True),\
   #                  StructField('Page size', IntegerType(), True)])

#1
class Log:
    def __init__(self,p, pt, ph, ps):
        self.p = p
        self.pt = pt
        self.ph = ph
        self.ps = ps
    def __repr__(self):
        return f"Log({self.p}, {self.pt}, {self.ph},{self.ps})"

#2
def convertStringToLog(string):
    p = string.split(" ")[0]
    pt = string.split(" ")[1]
    ph = int(string.split(" ")[2])
    ps = int(string.split(" ")[3])
  
    return Log(p,pt,ph,ps)


#3,4
#Log_pagecounts = lines.map(lambda x: [convertStringToLog(x).p,convertStringToLog(x).pt\
  #                              ,convertStringToLog(x).ph,convertStringToLog(x).ps] )

Log_pagecounts = lines.map(lambda x : convertStringToLog(x))
Log_pagecounts.cache
```

**Question 1**

```python
def print_record(logRdd):
  R = logRdd.map(lambda x : [x.p,x.pt,x.ph,x.ps] ).take(20)
  for i in range(20):
    print("Project code: "+R[i][0]+"\t Page:"+R[i][1]+"\t Page hits:"+str(R[i][2])+"\t Page size:"+str(R[i][3]))

print_record(Log_pagecounts)
```

**Question 2**

```python
def total_numOfRecords(logRdd):
  return logRdd.count()

total_numOfRecords(Log_pagecounts)
```

**Question 3**

```python
def max_pageSize(logRdd):
  return logRdd.map(lambda x : x.ps ).max()

def min_pageSize(logRdd):
  return logRdd.map(lambda x : x.ps ).min()


def avg_pageSize(logRdd):
  return logRdd.map(lambda x : x.ps ).sum()/total_numOfRecords(logRdd)

maxps= max_pageSize(Log_pagecounts)
minps= min_pageSize(Log_pagecounts)
avgps= avg_pageSize(Log_pagecounts)

print("max page size is: "+str(maxps))
print("min page size is: "+str(minps))
print("avg page size is: "+str(avgps))
```

**Question 4**

```python

pmax = Log_pagecounts.filter(lambda x: x.ps == maxps)\
                    .map(lambda x :"Project code: "+x.p+" ,Page: "+x.pt+" ,Page hits: "+str(x.ph) )

for i in pmax.collect(): print(i)

```

**Question 5**

```python
pmaxPop = Log_pagecounts.filter(lambda x: x.ps == maxps)\
                    .map(lambda x :"Project code: "+x.p+" ,Page: "+x.pt+" ,Page hits: "+str(x.ph) )

for i in pmaxPop.collect(): print(i)
```

**Question 6**

```python
fgPageSAvg = Log_pagecounts.filter(lambda x: x.ps > avgps)\
            .map(lambda x :"Project code: "+x.p+" ,Page: "+x.pt+" ,Page hits: "+str(x.ph) )

for i in fgPageSAvg.take(20): print(i)
```

**Question 7**

```python
pV = Log_pagecounts.map(lambda x: (x.p,x.ph)).reduceByKey(lambda x,y : x+y).map(lambda x: (x[1],x[0])).sortByKey(ascending=False)

for i in pV.take(10): print(i)
```

**Question 8**

```python
import string 
import re
def lower_clean_str(strr):
  return strr.translate(strr.maketrans('', '', string.punctuation)).lower()


count = Log_pagecounts.map(lambda x: x.pt).flatMap(lambda x: x.split('_') )\
                        .map(lambda x: lower_clean_str(x)).map(lambda word : (word,1))\
                        .filter( lambda x : re.match(r"^[a-z]+$",x[0]))\
                        .reduceByKey(lambda x,y : x+y)


for i in count.take(10): print(i)
```

```python
from collections import namedtuple
log = namedtuple("log", ["Project", "PageTitle", "PageHits", "PageSize"])

def StringToLog(s):
  C = s.split(" ")
  return log(C[0], C[1], int(C[2]), int(C[3]))

log_rdd = lines.map(StringToLog)
log_rdd.cache

log_rdd.take(20)
```

```python
from pyspark.sql.types import *
schema = StructType([StructField("Project", StringType(), True),\
                    StructField("title", StringType(), True),\
                     StructField("hits", LongType(), True),\
                   StructField("size", LongType(), False)])

df2 = log_rdd.toDF(schema)
df2.cache()
df2.printSchema()
df2.show(10)
```

**Question 3**

```python
import time 
start_time= time.time()
df2.createOrReplaceTempView("PROJECTS")
spark.sql("SELECT MAX(size) AS Max,MIN(size) AS Min, AVG(size) AS Avg FROM PROJECTS").show()

print(f"Execution time: {time.time()-start_time}")
#df.withColumn("age",col("age").cast(StringType()))
#from pyspark.sql.functions import max
#df2.cache()
#df2.select(max('size')).show()
#df2.select(min('size')).show()
#df2.select(mean('size')).show()
```

```python

import string 
def clean_str(strr):
  prestr = strr.translate(strr.maketrans('', '', string.punctuation)).lower()
  return prestr.translate(prestr.maketrans('', '', string.digits))

spark.udf.register("clean_str",clean_str)

start_time= time.time()

df_words = spark.sql("SELECT clean_str(word) As word,count(*) as occ FROM(SELECT explode(SPLIT(title,'_'))"\
+"  as word  FROM PROJECTS ) Group by word order by occ DESC").show(1,False)

print(f"Execution time: {time.time()-start_time}")
```
