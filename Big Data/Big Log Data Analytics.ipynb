```python
from pyspark.sql.types import *



#1
#schema = StructType([StructField('Project', StringType(), True),\
  #                   StructField('Page title', StringType(), True),\
  #                   StructField('Page hits', IntegerType(), True),\
   #                  StructField('Page size', IntegerType(), True)])

#1
class Log:
    def __init__(self,p, pt, ph, ps):
        self.p = p
        self.pt = pt
        self.ph = ph
        self.ps = ps
    def __repr__(self):
        return f"Log({self.p}, {self.pt}, {self.ph},{self.ps})"

#2
def convertStringToLog(string):
    p = string.split(" ")[0]
    pt = string.split(" ")[1]
    ph = int(string.split(" ")[2])
    ps = int(string.split(" ")[3])
  
    return Log(p,pt,ph,ps)


#3,4
#Log_pagecounts = lines.map(lambda x: [convertStringToLog(x).p,convertStringToLog(x).pt\
  #                              ,convertStringToLog(x).ph,convertStringToLog(x).ps] )

Log_pagecounts = lines.map(lambda x : convertStringToLog(x))


```

```python
def clean_str(strr):
  prestr = strr.translate(strr.maketrans('', '', string.punctuation)).lower()
  return prestr.translate(prestr.maketrans('', '', string.digits))


clean = udf(lambda z:clean_str(z),StringType())
df_words = df2.select(split(df2.title,"_").alias("words")) \
          .withColumn("value", lit(1))\
          .select(col("value"),explode(col('words')).alias("word"))\
          .withColumn('word',clean(col('word')))\
          .filter(col('word')!='')\
          .select(col('word'),col('value'))\
          .groupBy("word").sum("value")\
          .orderBy(col("sum(value)").desc())
          

df_words.printSchema()
df_words.show(50,False)
```
